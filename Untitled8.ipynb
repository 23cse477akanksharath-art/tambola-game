```python
# Tambola Game - EXTRA NESTED LOOPS VERSION

import random

# ========== FUNCTION 1: Get players ==========
def get_players():
    players = []
    num = int(input("How many players? "))
    
    # NESTED LOOP A: Get each player's numbers
    for p in range(num):
        print(f"\nPlayer {p+1} numbers:")
        nums = []
        
        # INNER LOOP: Get 5 numbers
        while len(nums) < 5:
            n = int(input(f"  Number {len(nums)+1}: "))
            if 1 <= n <= 20 and n not in nums:
                nums.append(n)
            else:
                print("Invalid or duplicate!")
        
        players.append({
            'id': p+1,
            'ticket': sorted(nums),
            'marked': []
        })
    return players

# ========== FUNCTION 2: Display tickets ==========
def show_tickets(players):
    print("\nTICKETS:")
    
    # NESTED LOOP B: Display all tickets
    for p in range(len(players)):  # Outer: players
        print(f"P{players[p]['id']}: ", end="")
        
        # INNER LOOP: Display numbers
        for n in range(len(players[p]['ticket'])):
            print(players[p]['ticket'][n], end=" ")
        print()

# ========== FUNCTION 3: Mark numbers ==========
def mark_numbers(players, called):
    marked_any = False
    
    # NESTED LOOP C: Check all players and numbers
    for p in range(len(players)):  # Level 1: players
        for n in range(len(players[p]['ticket'])):  # Level 2: numbers
            if players[p]['ticket'][n] == called:
                if called not in players[p]['marked']:
                    players[p]['marked'].append(called)
                    print(f"  âœ“ P{players[p]['id']} got it!")
                    marked_any = True
    return marked_any

# ========== FUNCTION 4: Show status ==========
def show_status(players):
    print("\nSTATUS:")
    
    # NESTED LOOP D: Show each player's progress
    for p in range(len(players)):  # Outer: players
        print(f"P{players[p]['id']}: ", end="")
        
        # INNER LOOP: Show marked numbers
        print("Marked[", end="")
        for m in range(len(players[p]['marked'])):
            print(players[p]['marked'][m], end=" ")
        print("] ", end="")
        
        # Show progress
        print(f"({len(players[p]['marked'])}/5)")

# ========== FUNCTION 5: Check winner ==========
def find_winner(players):
    # NESTED LOOP E: Check each player
    for p in range(len(players)):
        if len(players[p]['marked']) == 5:
            return players[p]
    return None

# ========== FUNCTION 6: Show final results ==========
def final_results(players, winner, called):
    print("\n" + "ğŸ†"*30)
    print(f"WINNER: Player {winner['id']}")
    print("ğŸ†"*30)
    
    print("\nFINAL SCORES:")
    
    # NESTED LOOP F: Show detailed results
    for p in range(len(players)):  # Level 1: players
        is_winner = " << WINNER" if players[p] == winner else ""
        print(f"\nPlayer {players[p]['id']}{is_winner}")
        print("  Ticket: ", end="")
        
        # Level 2: show each number with marking
        for n in range(len(players[p]['ticket'])):
            num = players[p]['ticket'][n]
            if num in players[p]['marked']:
                print(f"[{num}]", end=" ")
            else:
                print(f" {num} ", end=" ")
        print()
        
        print(f"  Marked: {sorted(players[p]['marked'])}")
        
        # Level 2 with Level 3: find missed numbers
        missed = []
        for n in range(len(players[p]['ticket'])):
            if players[p]['ticket'][n] not in players[p]['marked']:
                missed.append(players[p]['ticket'][n])
        print(f"  Missed: {missed}")

# ========== MAIN GAME ==========
def main():
    print("ğŸ² TAMBOLA GAME")
    print("="*30)
    
    # Step 1: Setup
    players = get_players()
    show_tickets(players)
    
    # Step 2: Game numbers
    numbers = list(range(1, 21))
    random.shuffle(numbers)
    called = []
    
    input("\nPress Enter to start...")
    
    # Step 3: Main game with nested loop
    winner = None
    for call_num in range(len(numbers)):  # Level 1: calls
        current = numbers[call_num]
        called.append(current)
        
        print(f"\nâ–¶ï¸ Call {call_num+1}: {current}")
        
        # Level 2 inside Level 1: mark numbers
        marked = mark_numbers(players, current)
        
        if not marked:
            print("  No one has it")
        
        # Level 2: show status
        show_status(players)
        
        # Level 2: check winner
        winner = find_winner(players)
        
        if winner:
            break
        
        if call_num < 19:
            input("Next? ")
    
    # Step 4: Results
    final_results(players, winner, called)

# Run the game
if __name__ == "__main__":
    main()
```

```python
# ============================================================================
# ğŸ² ULTIMATE TAMBOLA - The Most Exciting Number Game Ever! ğŸ²
# ============================================================================
# Game Features:
# âœ… Exciting animations and effects
# âœ… Multiple game modes
# âœ… Sound effects (text-based)
# âœ… Score tracking and statistics
# âœ… Player achievements
# âœ… Colorful console output
# ============================================================================

import random
import time
import os

# ============================================================================
# FUNCTIONS SECTION - All game functions
# ============================================================================

def clear_screen():
    """Clear the console screen for betterè§†è§‰æ•ˆæœ"""
    os.system('cls' if os.name == 'nt' else 'clear')

def print_slow(text, delay=0.03):
    """Print text with a typewriter effect"""
    for char in text:
        print(char, end='', flush=True)
        time.sleep(delay)
    print()

def show_banner():
    """Display game banner with animations"""
    clear_screen()
    banner = """
    â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
    â•‘  ğŸ¯   ğŸ¯   ğŸ¯   ULTIMATE TAMBOLA GAME   ğŸ¯   ğŸ¯   ğŸ¯        â•‘
    â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
    â•‘  âš¡ The Most Exciting Number Game in the Universe! âš¡        â•‘
    â•‘  ğŸ† 3 Game Modes â€¢ ğŸ® Multiplayer â€¢ ğŸ“Š Live Stats  ğŸ†       â•‘
    â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    """
    print("\033[96m" + banner + "\033[0m")  # Cyan color
    time.sleep(1)

def show_loading(message="Loading"):
    """Show loading animation"""
    print(f"\n{message}", end="")
    for i in range(3):
        time.sleep(0.5)
        print(".", end="", flush=True)
    print(" âœ…")

def get_player_count():
    """Get number of players with validation"""
    while True:
        try:
            print("\n" + "â•"*50)
            count = int(input("ğŸ® Enter number of players (1-6): "))
            if 1 <= count <= 6:
                return count
            else:
                print("âŒ Please enter between 1 and 6 players!")
        except ValueError:
            print("âŒ Invalid input! Enter a number.")

def choose_game_mode():
    """Let players choose game mode"""
    print("\n" + "â•"*50)
    print("ğŸ² GAME MODES:")
    print("â•"*50)
    print("1ï¸âƒ£  CLASSIC MODE     - First to complete all 5 numbers wins")
    print("2ï¸âƒ£  TOURNAMENT MODE  - Best of 3 rounds")
    print("3ï¸âƒ£  SPEED MODE       - 15 seconds per turn!")
    print("â•"*50)
    
    while True:
        try:
            mode = int(input("ğŸ¯ Choose mode (1-3): "))
            if 1 <= mode <= 3:
                return mode
        except ValueError:
            pass
        print("âŒ Please enter 1, 2, or 3!")

def create_player_ticket(player_num):
    """Create one player's ticket with exciting interface"""
    print(f"\n{'â•'*50}")
    print(f"ğŸ® PLAYER {player_num} - CREATE YOUR DREAM TICKET")
    print(f"{'â•'*50}")
    print("ğŸ¯ Pick 5 lucky numbers (1-20)")
    print("ğŸ’¡ Tip: Choose numbers that feel lucky!\n")
    
    player_numbers = []
    
    while len(player_numbers) < 5:
        try:
            num = int(input(f"  ğŸ² Number {len(player_numbers)+1}: "))
            
            if num < 1 or num > 20:
                print("  âŒ Number must be 1-20!")
            elif num in player_numbers:
                print("  âŒ You already have this number!")
            else:
                player_numbers.append(num)
                print(f"  âœ… Added! {5 - len(player_numbers)} more to go")
                
                # Show current ticket
                if player_numbers:
                    print(f"     Current: {sorted(player_numbers)}")
        except ValueError:
            print("  âŒ Invalid! Enter a number.")
    
    # Celebration effect
    print(f"\nâœ¨ Player {player_num} ticket created! âœ¨")
    print(f"   ğŸ« Your lucky numbers: {sorted(player_numbers)}")
    time.sleep(1)
    
    return sorted(player_numbers)

def create_all_players(num_players):
    """Create tickets for all players"""
    print(f"\n{'â˜…'*60}")
    print("ğŸŒŸ TICKET CREATION PHASE ğŸŒŸ")
    print(f"{'â˜…'*60}")
    
    all_players = []
    
    for i in range(num_players):
        player_ticket = create_player_ticket(i + 1)
        
        player_data = {
            'id': i + 1,
            'ticket': player_ticket,
            'marked': [],
            'score': 0,
            'rounds_won': 0,
            'achievements': []
        }
        all_players.append(player_data)
        
        if i < num_players - 1:
            input("\nğŸ‘‰ Press Enter for next player...")
    
    return all_players

def display_all_tickets(players):
    """Show all tickets in a beautiful format"""
    print(f"\n{'ğŸ«'*30}")
    print("ğŸŒŸ ALL PLAYER TICKETS")
    print(f"{'ğŸ«'*30}")
    
    for player in players:
        # Create visual ticket
        print(f"\nâ”Œâ”€ PLAYER {player['id']} ", end="")
        print("â”€" * 25 + "â”")
        print("â”‚ ", end="")
        
        # Display numbers with formatting
        for num in player['ticket']:
            if num in player['marked']:
                print(f"\033[92m[{num:2}]\033[0m ", end="")  # Green for marked
            else:
                print(f" {num:2}  ", end="")
        
        print(" â”‚")
        print(f"â”‚ {'âœ“' * len(player['marked'])}{'â—‹' * (5 - len(player['marked']))}        â”‚")
        print(f"â””{'â”€' * 35}â”˜")

def draw_number_animation():
    """Animation for drawing number"""
    print("\nğŸ² Drawing number", end="")
    for i in range(3):
        time.sleep(0.3)
        print(".", end="", flush=True)
    time.sleep(0.3)

def check_number_in_ticket(player, called_num):
    """Check if number is in ticket with excitement"""
    for i, num in enumerate(player['ticket']):
        if num == called_num and num not in player['marked']:
            player['marked'].append(num)
            player['score'] += 10
            return True
    return False

def check_bingo(players):
    """Check if anyone has BINGO"""
    for player in players:
        if len(player['marked']) == 5 and player not in getattr(check_bingo, "winners", []):
            if not hasattr(check_bingo, "winners"):
                check_bingo.winners = []
            check_bingo.winners.append(player)
            return player
    return None

def show_live_status(players, called_numbers, round_num=1):
    """Show exciting live status"""
    print(f"\n{'ğŸ“Š'*30}")
    print(f"ROUND {round_num} LIVE STATUS")
    print(f"{'ğŸ“Š'*30}")
    
    print(f"ğŸ“¢ Called: ", end="")
    for num in called_numbers:
        print(f"{num} ", end="")
    print(f"({len(called_numbers)} numbers)")
    
    print(f"\n{'â”€'*60}")
    print(f"{'Player':<8} {'Progress':<20} {'Marked':<15} {'Score':<8}")
    print(f"{'â”€'*60}")
    
    for player in players:
        # Progress bar
        marked = len(player['marked'])
        progress = "â–ˆ" * marked + "â–‘" * (5 - marked)
        
        # Player row
        print(f"P{player['id']:<7} [{progress:<5}] {marked}/5   ", end="")
        print(f"{sorted(player['marked']):<15} {player['score']:<8}")

def show_exciting_number(number):
    """Show the called number with excitement"""
    print(f"\n{'âœ¨'*20}")
    print(f"     ğŸ¯ NUMBER {number} IS CALLED! ğŸ¯")
    print(f"{'âœ¨'*20}")

def check_achievements(player):
    """Check and award achievements"""
    achievements = []
    
    if len(player['marked']) == 1 and player['marked'][0] == player['ticket'][0]:
        achievements.append("ğŸ”¥ FIRST BLOOD")
        player['score'] += 5
    
    if len(player['marked']) == 5:
        achievements.append("ğŸ† BINGO CHAMPION")
        player['score'] += 50
    
    if len(player['marked']) >= 3 and len([n for n in player['marked'] if n > 10]) == 3:
        achievements.append("ğŸ’ª HIGH ROLLER")
        player['score'] += 15
    
    return achievements

def show_round_results(players, round_winner):
    """Show round results with excitement"""
    print(f"\n{'ğŸ†'*50}")
    print(f"      ROUND WINNER: PLAYER {round_winner['id']}! ğŸ†")
    print(f"{'ğŸ†'*50}")
    
    print(f"\nğŸ“Š ROUND STATISTICS:")
    for player in players:
        print(f"  Player {player['id']}: {len(player['marked'])}/5 marked | Score: {player['score']}")

def show_final_results(players):
    """Show grand finale results"""
    print(f"\n{'ğŸ‰'*50}")
    print("           GRAND FINALE - TOURNAMENT WINNER")
    print(f"{'ğŸ‰'*50}")
    
    # Sort by score
    sorted_players = sorted(players, key=lambda x: x['score'], reverse=True)
    
    print(f"\n{'â”€'*60}")
    print(f"{'Rank':<6} {'Player':<10} {'Score':<10} {'Marked':<10} {'Achievements':<15}")
    print(f"{'â”€'*60}")
    
    medals = ["ğŸ¥‡ GOLD", "ğŸ¥ˆ SILVER", "ğŸ¥‰ BRONZE"]
    for i, player in enumerate(sorted_players):
        if i < 3:
            rank = medals[i]
        else:
            rank = f"{i+1}th"
        
        # Get achievements for display
        achievements = ", ".join(player.get('achievements', []))[:15]
        
        print(f"{rank:<6} P{player['id']:<9} {player['score']:<10} {len(player['marked'])}/5       {achievements}")
    
    print(f"\n{'ğŸŠ'*50}")
    print(f"    ğŸ† CHAMPION: PLAYER {sorted_players[0]['id']} WITH {sorted_players[0]['score']} POINTS! ğŸ†")
    print(f"{'ğŸŠ'*50}")

# ============================================================================
# MAIN GAME CLASS
# ============================================================================

class TambolaGame:
    def __init__(self):
        self.players = []
        self.called_numbers = []
        self.round_number = 1
        self.game_mode = 1
        self.numbers_pool = list(range(1, 21))
    
    def reset_round(self):
        """Reset for new round"""
        for player in self.players:
            player['marked'] = []
        self.called_numbers = []
        self.numbers_pool = list(range(1, 21))
        random.shuffle(self.numbers_pool)
    
    def play_round(self):
        """Play one round of Tambola"""
        print(f"\n{'ğŸ”¥'*50}")
        print(f"           ROUND {self.round_number} START!")
        print(f"{'ğŸ”¥'*50}")
        
        round_winner = None
        call_count = 0
        
        # Show initial tickets
        display_all_tickets(self.players)
        input("\nğŸ‘‰ Press Enter to start calling numbers...")
        
        # Main round loop
        while not round_winner and self.numbers_pool:
            call_count += 1
            clear_screen()
            
            # Draw number with animation
            draw_number_animation()
            current_number = self.numbers_pool.pop(0)
            self.called_numbers.append(current_number)
            
            show_exciting_number(current_number)
            time.sleep(1)
            
            # Check all players
            marked_this_round = False
            for player in self.players:
                if check_number_in_ticket(player, current_number):
                    print(f"\nâœ… Player {player['id']} got it!")
                    marked_this_round = True
                    
                    # Check for achievements
                    new_achievements = check_achievements(player)
                    if new_achievements:
                        player['achievements'].extend(new_achievements)
                        print(f"   ğŸ† Achievement: {new_achievements[0]}")
                    time.sleep(0.5)
            
            if not marked_this_round:
                print(f"\nâŒ No one had {current_number}")
            
            # Show live status
            show_live_status(self.players, self.called_numbers, self.round_number)
            
            # Check for BINGO
            round_winner = check_bingo(self.players)
            
            if round_winner:
                print(f"\n{'ğŸŠ'*40}")
                print(f"     BINGO! Player {round_winner['id']} WINS THE ROUND! ğŸŠ")
                print(f"{'ğŸŠ'*40}")
                round_winner['rounds_won'] += 1
            else:
                if self.game_mode == 3:  # Speed mode - 3 seconds per turn
                    time.sleep(1.5)
                else:
                    input("\nğŸ‘‰ Press Enter for next number...")
        
        return round_winner
    
    def play_game(self):
        """Main game controller"""
        show_banner()
        
        # Game setup
        self.game_mode = choose_game_mode()
        show_loading("Setting up game")
        
        num_players = get_player_count()
        self.players = create_all_players(num_players)
        
        show_loading("Initializing game engine")
        
        # Tournament mode handling
        if self.game_mode == 2:  # Tournament mode
            target_wins = 2 if num_players <= 3 else 1
            print(f"\nğŸ† TOURNAMENT MODE - First to win {target_wins} rounds wins!")
            time.sleep(2)
            
            tournament_winner = None
            while not tournament_winner:
                round_winner = self.play_round()
                show_round_results(self.players, round_winner)
                
                # Check tournament winner
                for player in self.players:
                    if player['rounds_won'] >= target_wins:
                        tournament_winner = player
                        break
                
                if not tournament_winner:
                    self.round_number += 1
                    self.reset_round()
                    input("\nğŸ‘‰ Press Enter for next round...")
            
            print(f"\n{'ğŸ†'*60}")
            print(f"     TOURNAMENT CHAMPION: PLAYER {tournament_winner['id']}! ğŸ†")
            print(f"{'ğŸ†'*60}")
        
        elif self.game_mode == 3:  # Speed mode
            print("\nâš¡ SPEED MODE ACTIVATED! Quick turns! âš¡")
            time.sleep(1)
            round_winner = self.play_round()
            show_round_results(self.players, round_winner)
        
        else:  # Classic mode
            round_winner = self.play_round()
            show_round_results(self.players, round_winner)
        
        # Show final results
        show_final_results(self.players)
        
        # Play again?
        print("\n" + "â•"*50)
        again = input("ğŸ® Play again? (yes/no): ").lower()
        return again == 'yes'

# ============================================================================
# GAME START
# ============================================================================

if __name__ == "__main__":
    playing = True
    while playing:
        game = TambolaGame()
        playing = game.play_game()
        
        if playing:
            print("\nğŸ”„ Restarting game...")
            time.sleep(2)
    
    print(f"\n{'ğŸ‘‹'*40}")
    print("     Thanks for playing ULTIMATE TAMBOLA! See you next time!")
    print(f"{'ğŸ‘‹'*40}")
```
